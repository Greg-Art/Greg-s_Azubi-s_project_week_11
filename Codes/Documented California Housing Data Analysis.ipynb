# Feature Engineering

```python
housing['rooms_per_household']= housing['total_rooms']/housing['households']
housing["bedrooms_per_room"]= housing["total_bedrooms"]/housing["total_rooms"]
housing["population_per_household"]= housing["population"]/housing["households"]
```

# Data Splitting 

```python
## importing all our sklearn modules

from sklearn.model_selection import train_test_split

from sklearn.preprocessing import OneHotEncoder

from sklearn.preprocessing import RobustScaler

from sklearn.impute import SimpleImputer

from sklearn.pipeline import Pipeline

from sklearn.compose import ColumnTransformer

from sklearn.neighbors import KNeighborsRegressor
```

```python
train,test= train_test_split(housing, train_size= 0.85, test_size= 0.15, random_state= 35)
```

Let's get our features and target

```python
train_features= train.drop("median_house_value", axis= 1)

num_train_features= train_features.drop("ocean_proximity",axis= 1)

cat_train_features= train.ocean_proximity

train_target= train.median_house_value

test_features= test.drop("median_house_value", axis= 1)

num_test_features= test_features.drop("ocean_proximity",axis= 1)

cat_test_features= test.ocean_proximity

test_target= test.median_house_value
```

creating a pipeline

```python
num_pipeline= Pipeline([("Impute", SimpleImputer()),("Scaler", RobustScaler())])

num_attr= list(num_train_features)

cat_attr= ["ocean_proximity"]

full_pipeline= ColumnTransformer([("num", num_pipeline, num_attr),("cat", OneHotEncoder(), cat_attr)])
```

```python
full_pipeline.fit(train_features)
```

```python
train_features= pd.DataFrame(full_pipeline.transform(train_features))

test_feature= pd.DataFrame(full_pipeline.transform(test_features))
```

```python
train_features
```

```python
model= KNeighborsRegressor()
```

```python
model.fit(train_features, train_target)
```

```python
pred=model.predict(train_features)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
error= mean_absolute_error(train_target, pred)
```

```python
(error/train_target.mean())*100
```

```python
actual_pred= model.predict(test_feature)

act_error= mean_absolute_error(test_target, actual_pred)

(act_error/test_target.mean())*100
```
